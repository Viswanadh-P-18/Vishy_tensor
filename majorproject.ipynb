# AUTOMATING PREDICTIONS

```python
def textfit(X,Y,ml_model,nlp_model):
    X_c=nlp_model.fit_transform(X)
    x_train,x_test,y_train,y_test=train_test_split(X_c,y)
    ml=ml_model.fit(x_train,y_train)
    acc=ml_model.score(x_test,y_test)
    print(acc)
    
    w=c.get_feature_names()
    coef=ml.coef_.tolist()[0]
    coef_df=pd.DataFrame({'Coeff':coef,'Words':w})
    coef_df=coef_df.sort_values(['Coeff','Words'],ascending=False)
    print("The top 20 words :")
    print("\n")
    print(coef_df.head(20))
    print("The below 20 words :")
    print("\n")
    print(coef_df.tail(20))
    
```

# SENTIMENT ANALYSIS USING TENSORFLOW

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

# Input layer and one hidden layer (you can add more layers and units as needed)
model.add(Dense(16, input_dim=x_train.shape[1], activation='relu'))  # Hidden layer
model.add(Dense(1, activation="sigmoid"))
```

```python
X
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
model.compile(optimizer="Adagrad", loss='binary_crossentropy')
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10)
```

```python
losses_df=pd.DataFrame(model.history.history)
```

```python
losses_df.plot(kind="line")
```

```python
def predict_sentiment(texts):
    text=pd.Series(texts)
    x_array=c.transform(text)
    prediction=model.predict(x_array)
    if prediction[0]>=0.5:
        return "This is a positive statement"
    else:
        return "This is a negative statement"
```
