```python
data=pd.read_csv("loan.csv")
```

```python
data.replace({"Self_Employed":{"No":0,"Yes":1}},inplace=True)
```

```python
data["Property_Area"].value_counts()
```

```python
data.replace({"Property_Area":{"Semiurban":2,"Urban":1,"Rural":0}},inplace=True)
```

```python
data.head(5)
```

```python
data["Gender"].value_counts()
```

```python
data.replace({"Gender":{"Male":1,"Female":0}},inplace=True)
```

```python
data.replace({"Married":{"Yes":1,"No":0}},inplace=True)
```

```python
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
```

```python
data.shape
```

```python
x=data.drop("Loan_Status",axis=1)
y=data["Loan_Status"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
model.add(Dense(72,activation="relu"))
model.add(Dense(36,activation="relu"))
model.add(Dense(18,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(6,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
```

```python
sc.fit_transform(x_train)
```

```python
sc.fit(x_test)
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=100)
```

```python
losses_df=pd.DataFrame(model.history.history)
```
